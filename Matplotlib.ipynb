### Matplotlib used for data visualization

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=np.arange(0,10)
y=np.arange(10,20)
```

```python
plt.scatter(x,y,c='b')
#plt.figure(figsize=(6,8))
plt.title('X~Y Graph')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.savefig('X_Y.png')
```

```python
y=x*x
```

```python
print(plt.plot(x,y,'b'))
plt.show()
print(plt.plot(x,y,'b*-'))
plt.show()
print(plt.plot(x,y,'b--'))
plt.show()
print(plt.plot(x,y,'bo'))
plt.show()
print(plt.plot(x,y,'b*',linestyle='dashed',linewidth=2,markersize=12))
plt.show()
```

### Subplots for defining the number of rows and columns for graphs in same CANVAS

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r*--')
plt.subplot(2,2,2)
plt.plot(x,y,'b*--')
plt.subplot(2,2,3)
plt.plot(x,y,'g*--')
plt.subplot(2,2,4)
plt.plot(x,y,'y--')
```

### Drawing line equation

```python
#m=Slope
#c=intercept
m=2
c=3
x=np.arange(0,10)
y=m*x+c
```

```python
plt.plot(x,y,'r')
plt.title('Line')
plt.xlabel('X Value')
plt.ylabel('Y Label')
plt.grid()
plt.show()
```

### Sin(x) curve

```python
pi=np.pi
```

```python
pi
```

```python
x=np.arange(0,4*pi,0.1)
```

```python
y_sin=np.sin(x)
```

```python
y_cos=np.cos(x)
```

```python
plt.plot(x,y_sin,'b')
plt.title('Sin(X)')
plt.xlabel('X')
plt.ylabel('Sin(x) Value')
```

```python
plt.plot(x,y_cos,'b')
plt.title('Cos(X)')
plt.xlabel('X')
plt.ylabel('Cos(x) Value')
```

### Histogram 
#A histogram shows the frequency on the vertical axis and the horizontal axis is another dimension. Usually it has bins, where #every bin has a minimum and maximum value. Each bin also has a frequency between x and infinite.

```python
x = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]

num_bins = 5
```

```python
plt.hist(x,bins=num_bins,color='b')
plt.title('Histogram')
plt.xlabel('X-BINS')
plt.ylabel('Y-Frequency')
plt.show()
```

```python
print(np.sort(x))
```

```python
plt.boxplot(x,showmeans=True,vert=False,patch_artist=True)
plt.title('BOX PLOT')
```

### PIE Plot

```python
labels=['USA','Spain','Italy','France']
colors=['r','b','g','gold']
data=[849092,21302,187327,159877]
explode=[0.05,0,0,0]
```

```python
plt.pie(data,labels=labels,colors=colors,explode=explode,autopct='%1.1f%%',shadow=True)
plt.show()
```
