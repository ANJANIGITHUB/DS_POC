### BiVariate Analysis

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg',color='g')
```

```python
sns.pairplot(df)
```

```python
df['sex'].value_counts()
```

```python
sns.pairplot(df,hue='sex')
```

```python
df['smoker'].value_counts()
```

```python
sns.pairplot(df,hue='smoker')
```

### Distplot used to check the distribution of any feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
df.groupby(by='tip').count().head(10)
```

### Plots for Categorical Features

### countplot

```python
sns.countplot(x='sex',data=df)
#plt.grid()
```

```python
sns.countplot(x='sex',hue='smoker',data=df)
plt.grid()
```

```python
print(sns.countplot(x='day',data=df))
plt.show()
print(sns.countplot(y='day',data=df))
plt.show()
```

```python
sns.barplot(x='total_bill',y='smoker',data=df)
```

```python
sns.barplot(y='total_bill',x='sex',data=df)
```

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
df.groupby(by=['sex'])['total_bill'].mean()
```

```python
sns.boxplot(data=df,orient='v')
```
