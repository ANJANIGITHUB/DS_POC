```python
from sklearn.datasets import load_iris
import sklearn.tree as tr
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import sklearn
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
print(sklearn.__version__)
```

```python
iris_data=load_iris()
```

```python
iris_data.keys()
```

```python
print(iris_data['DESCR'])
```

```python
X=pd.DataFrame(iris_data['data'],columns=iris_data['feature_names'])
```

```python
X.head()
```

```python
y=pd.Categorical.from_codes(iris_data['target'],iris_data['target_names'])
```

```python
y
```

```python
le=LabelEncoder()
```

```python
y_enc=le.fit_transform(y)
```

```python
y_enc
```

```python
rf=RandomForestClassifier(n_estimators=20,max_depth=10,criterion='gini',min_samples_leaf=8)
```

```python
rf.fit(X,y_enc)
```

### Visualization of Random Forest Classifier

```python
#This is showing list of 20 trees which we have told to create while training the Random Forest in previous steps.
rf.estimators_
```

```python
#First Tree
plt.figure(figsize=(12,8))
tr.plot_tree(rf.estimators_[1],filled=True)
```

```python
#Second Tree
plt.figure(figsize=(12,8))
tr.plot_tree(rf.estimators_[2],filled=True)
```

```python
#Simillarly we can generate all the trees mention in random forest upto n_estimators parameter size
#Fifth Tree
plt.figure(figsize=(12,8))
tr.plot_tree(rf.estimators_[5],filled=True)
```

### Random Forest Classifier Text

```python
for i in rf.estimators_:
    print(tr.export_text(i))
```
